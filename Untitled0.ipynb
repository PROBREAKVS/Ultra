```python
texts = [
    "наша таня громко плачет уронила в речку мячик тише танечка не плачь не утонет в речке мячь",
    "идет бычок качается вздыхает на ходу ох доска кончается сейчас я упаду",
    "ох напрасно мы решили покатать кота в машине кот кататься не привык опракинул грузовик",
    "зайку бросила хозяйка под дождем остался зайка со скаймеки слезть не смог весь до ниточки промок",
    "я люблю свою лошадку причешу ей шерстку гладко гребешком приглажу хвостик и верхом поеду в гости"
]
```

```python
texts
```

```python
import pandas as pd

pairs = []

for text in texts:
  words = text.split()
  for i in range(0, len(words) - 1):
    pairs.append((words[i], words[i+1]))

df = pd.DataFrame(pairs, columns=["Контекст", "Следующее слово"])

df

unique_words = sorted(set(df["Контекст"]) | set(df["Следующее слово"]))

frequency_matrix = pd.DataFrame(0, index=unique_words, columns=unique_words)

for _, row in df.iterrows():
  context, next_word = row['Контекст'], row['Следующее слово']
  frequency_matrix.loc[context, next_word] += 1

probability_matrix = frequency_matrix.div(frequency_matrix.sum(axis=1), axis=0).fillna(0)
probability_matrix
```

```python
def predict_next_word(context):
  if context in probability_matrix.index:
    probabilities = probability_matrix.loc[context]
    return probabilities.sample(weights=probabilities).idxmax()
  else:
    return "SUS"
context = "бычок"

next_word = predict_next_word(context)
print(next_word)
next_word = predict_next_word(next_word)
print(next_word)
next_word = predict_next_word(next_word)
print(next_word)
next_word = predict_next_word(next_word)
print(next_word)
next_word = predict_next_word(next_word)
print(next_word)
next_word = predict_next_word(next_word)
print(next_word)
next_word = predict_next_word(next_word)
print(next_word)

```
